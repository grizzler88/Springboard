### Import sqlite3 and setting up connection

```python
import sqlite3
from sqlite3 import Error
```

```python
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn
```

```python
database = "sqlite_db_pythonsqlite.db"

conn = create_connection(database)
```

### Question 10

Produce a list of facilities with a total revenue less than 1000. The output of facility name and total revenue, sorted by revenue. Remember that there's a different cost for guests and members!

```python
pd.read_sql_query("""
SELECT facility, sum( revenue ) AS revenue

FROM ( SELECT f.name AS facility,
CASE WHEN memid =0 THEN guestcost
ELSE membercost END AS revenue
FROM Bookings AS b
LEFT JOIN Facilities AS f
USING ( facid )
LEFT JOIN Members AS m
USING ( memid )
) AS sub

WHERE revenue <1000
GROUP BY facility
ORDER BY revenue DESC 
""", 
                  
conn)
```

### Question 11

Produce a report of members and who recommended them in alphabetic surname,firstname order.

```python
pd.read_sql_query(

"""SELECT m1.firstname, m1.surname, m2.firstname, m2.surname
FROM Members AS m1
LEFT JOIN Members AS m2 ON m1.recommendedby = m2.memid
ORDER BY m1.surname, m1.firstname;"""
,
conn)
```

### Question 12

Find the facilities with their usage by member, but not guests.

```python
pd.read_sql_query(
    
"""SELECT f.name AS facility, sum(slots) AS usage
FROM Bookings AS m
INNER JOIN Facilities AS f
USING (facid)
WHERE memid<>0
GROUP BY facility
ORDER BY usage;"""
,
conn)
```

### Question 13

Find the facilities usage by month, but not guests.

```python
pd.read_sql_query(

'''SELECT f.name AS facility, strftime('%m', b.starttime) AS month, sum(slots) AS usage 
FROM Bookings AS b
LEFT JOIN Facilities AS f
USING (facid)
WHERE memid<>0
GROUP BY facility, month
ORDER BY month, usage
'''
,
conn)
```
